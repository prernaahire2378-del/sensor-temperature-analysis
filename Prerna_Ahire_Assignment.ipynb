```python
import numpy as np

data = np.array([[22.5,19.0,31.2,28.7,25.1],
                 [17.3,22.8,30.5,26.4,21.9],
                 [33.1,29.6,18.4,24.0,27.8],
                 [20.2,23.5,31.9,28.1,22.6]])

print(data)
```

# Task 1 – Print Shape and Mean per Station

```python
# Print shape
print("Shape of data:", data.shape)

# Mean temperature per station
mean_temp = np.mean(data, axis=1)

print("Mean temperature per station:", mean_temp)
```

Explanation (simple):

(4,5) → 4 stations and 5 readings each

axis=1 → calculate mean for each row (station)

# Task 2 – Extract Temperatures Above 28°C

```python
# Boolean mask for temperatures above 28
high_temp = data[data > 28]

print("Temperatures above 28°C:", high_temp)
```

Explanation:

data > 28 finds values greater than 28

It returns them as a 1D array

# Task 3 – Normalize Data 

```python


normalized = (data - data.min()) / (data.max() - data.min())

# round to 2 decimal places
normalized = np.round(normalized, 2)

print("Normalized Data:\n", normalized)
```

Explanation:

min value = 17.3

max value = 33.1

Converts all values between 0 and 1
